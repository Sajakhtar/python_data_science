# Cross Validation Bias Variance Tradeoff

Bias Variance Trade Off is a fundamental topic of understanding your model's performance.

Review Chapter 2 of Introduction to Statistical Learning (Gareth James et al.)

http://faculty.marshall.usc.edu/gareth-james/ISL/ISLR%20Seventh%20Printing.pdf


* The bias-variance trade-off is the point where we are adding noise by adding model complexity (flexibility)
* The training error goes down as it has to, but the test error start going up
* The model after bias trade-off begins to **overfit**

### Bias-variance trade-off Diagram
* each hit on the target is a model
    * repeat model, varying the params and training data
* hit on the Bullseye = model predicts perfectly
* each hit represents an individual realization of our model, given chance variability in training data
* sometimes we get a good distribution of training data so we predict very well and are close to the bullseye
* other times the training data may be full of outliers or non-standard values resulting in poor predictions
* different realizations of the model results in a scatter of hits on the target
* we aim for low bias and low variance
* realistically you'll have to trade off variance or bias


* common temptation for beginners is to continually add complexity/flexiblity to a model until it fits the training set very well (overfitting training data)
* But then model fails to predict the test data well enough (large errors on test data)

* You want to balance out bias and variance of your model to the point where your test data and training data have reached some sort of minimum and grouping toegether.
