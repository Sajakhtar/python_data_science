## Introduction

Matplotlib is the "grandfather" library of data visualization with Python. It was created by John Hunter. He created it to try to replicate MatLab's (another programming language) plotting capabilities in Python. So if you happen to be familiar with matlab, matplotlib will feel natural to you.

It is an excellent 2D and 3D graphics library for generating scientific figures. 

Some of the major Pros of Matplotlib are:

* Generally easy to get started for simple plots
* Support for custom labels and texts
* Great control of every element in a figure
* High-quality output in many formats
* Very customizable in general

Matplotlib allows you to create reproducible figures programmatically. Let's learn how to use it! Before continuing this lecture, I encourage you just to explore the official Matplotlib web page: http://matplotlib.org/

# Matplotlib Object Oriented Method
Now that we've seen the basics, let's break it all down with a more formal introduction of Matplotlib's Object Oriented API. This means we will instantiate figure objects and then call methods or attributes from that object.

## Introduction to the Object Oriented Method

The main idea in using the more formal Object Oriented method is to create figure objects and then just call methods or attributes off of that object. This approach is nicer when dealing with a canvas that has multiple plots on it. 

To begin we create a figure instance. Then we can add axes to that figure:

```python
# OOP method - call figure objects and call methods off of them
fig = plt.figure() # blank canvas

# Add set of axes to figure
axes = fig.add_axes([0.1,0.1,0.8,0.8]) # pass in a list [left%, bottom%, width%, height%]  %of canvas (range 0 to 1)

# Plot on that set of axes
axes.plot(x,y, 'bo-')
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title("Plot title")
```

Code is a little more complicated, but the advantage is that we now have full control of where the plot axes are placed, and we can easily add more than one axis to the figure:

```python
# Creates blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) # Main axes [left%, bottom%, width%, height%]  %of canvas (range 0 to 1)
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) # inset axes

# Larger Figure Axes 1
axes1.plot(x,y, 'bo-')
axes1.set_xlabel('X label')
axes1.set_ylabel('Y label')
axes1.set_title("Main Plot")

# Insert Figure Axes 2
axes2.plot(y,x, 'r--')
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title("Inset Plot")
```

```python
# Creates blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) # Main axes [left%, bottom%, width%, height%]  %of canvas (range 0 to 1)
axes2 = fig.add_axes([0.65,0.22,0.23,0.16]) # inset axes

# Main Figure Axes 1 - add plot
axes1.plot(x,y, 'bo-')
axes1.set_xlabel('X label')
axes1.set_ylabel('Y label')
axes1.set_title("Main Plot")

# Insert Figure Axes 2 - add plot
axes2.plot(y,x, 'r--')
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title("Inset Plot")
```

## subplots()

The plt.subplots() object will act as a more automatic axis manager.

Basic use cases:

```python
# Use similar to plt.figure() except use tuple unpacking to grab fig and axes
# tuple unpacking
fig,axes = plt.subplots()

# Now use the axes object to add stuff to plot
axes.plot(x,y,'g--')
```

```python
# Empty canvas of 1 by 2 subplots
fig, axes = plt.subplots(nrows=3, ncols=4) # 1,1 by default

plt.tight_layout() # to stop overlapping
```

```python
# Empty canvas of 1 by 2 subplots
fig, axes = plt.subplots(nrows=1, ncols=4) # 1,1 by default

plt.tight_layout() # to stop overlapping
```

```python
# Axes is an array of axes to plot on
# can iterate through it
axes
```

```python
for ax in axes:
    ax.plot(x,y,'g--')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

fig
```

```python
# Empty canvas of 1 by 2 subplots
fig, axes = plt.subplots(nrows=4, ncols=5) # 1,1 by default

plt.tight_layout() # to stop overlapping

for axis in axes:
    for ax in axis:
        ax.plot(x,y,'g--')
        ax.set_xlabel('x')
        ax.set_ylabel('y')
        ax.set_title('title')
```

```python
# Empty canvas of 1 by 2 subplots
fig, axes = plt.subplots(nrows=1, ncols=2) # 1,1 by default

# since axes is a list we can index and plot
axes[0].plot(x,y, 'go-')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('title1')

axes[1].plot(y,x,'r--')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].set_title('title2')

plt.tight_layout() # to stop overlapping
```

### Figure size, aspect ratio and DPI

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the `figsize` and `dpi` keyword arguments. 
* `figsize` is a tuple of the width and height of the figure in inches
* `dpi` is the dots-per-inch (pixel per inch). 

For example:

```python
fig = plt.figure(figsize=(3,2), dpi=100) # usually leave DPI to default value of None

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,'--')
```

```python
fig = plt.figure(figsize=(8,2), dpi=100) # usually leave DPI to default value of None

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,'--')
```

The same arguments can also be passed to layout managers, such as the `subplots` function:

```python
fig,axes = plt.subplots(nrows=1, ncols=2, figsize=(12,3))

# since axes is a list we can index and plot
axes[0].plot(x,y, 'go-')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('title1')

axes[1].plot(y,x,'r--')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].set_title('title2')

plt.tight_layout() # to stop overlapping
```

## Legends, labels and titles

Now that we have covered the basics of how to create a figure canvas and add axes instances to the canvas, let's look at how decorate a figure with titles, axis labels, and legends.

**Figure titles**

A title can be added to each axis instance in a figure. To set the title, use the `set_title` method in the axes instance:

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2, label='X^2')
ax.plot(x,x**2.5, label='X^2.5')
ax.plot(x,x**3, label='X^3')


ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
#ax.legend(loc='best') #loc= best 0, upper left 2, lower right 4, center 10, center left 6, etc..
#ax.legend(loc=0)
ax.legend(loc=(0.05,0.2)) # specify tuple (left, bottom)
```

Notice how are legend overlaps some of the actual plot!

## Setting colors, linewidths, linetypes

Matplotlib gives you *a lot* of options for customizing colors, linewidths, and linetypes. 

There is the basic MATLAB like syntax (which I would suggest you avoid using for more clairty sake:

### Colors with MatLab like syntax

With matplotlib, we can define the colors of lines and other graphical elements in a number of ways. First of all, we can use the MATLAB-like syntax where `'b'` means blue, `'g'` means green, etc. The MATLAB API for selecting line styles are also supported: where, for example, 'b.-' means a blue line with dots:

```python
# MATLAB style line color and style 
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y*2, 'go--')
ax.plot(x,x**2, color='purple')
ax.plot(x,x**2.5, color='#FF8C00') # hex code
ax.plot(x,y*1.5, 'black', alpha=0.5) # half-transparant
```

### Line and marker styles

To change the line width, we can use the `linewidth` or `lw` keyword argument. The line style can be selected using the `linestyle` or `ls` keyword arguments:

```python
fig,ax = plt.subplots(figsize=(12,6))

# linewidth= or lw=
ax.plot(x, x+1, color="violet", linewidth=0.25)
ax.plot(x, x+2, color="violet", linewidth=0.5)
ax.plot(x, x+3, color="violet", linewidth=1)
ax.plot(x, x+4, color="violet", linewidth=2)
ax.plot(x, x+5, color="violet", linewidth=4)

# possible linestype options ‘-‘, ‘–’, ‘-.’, ‘:’, ‘steps’
# linestyle= or ls=
ax.plot(x, x+6, color="green", lw=3, linestyle='-')
ax.plot(x, x+7, color="green", lw=3, ls='-.')
ax.plot(x, x+8, color="green", lw=3, ls=':')
ax.plot(x, x+9, color="green", lw=3, ls='--')

# Markers
# marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+10, color="blue", lw=1, ls='-', marker='+')
ax.plot(x, x+11, color="blue", lw=1, ls='-', marker='o')
ax.plot(x, x+12, color="blue", lw=1, ls='-', marker=',')
ax.plot(x, x+13, color="blue", lw=1, ls='-', marker='.')
ax.plot(x, x+14, color="blue", lw=1, ls='-', marker='s')
ax.plot(x, x+15, color="blue", lw=1, ls='-', marker='1')
ax.plot(x, x+16, color="blue", lw=1, ls='-', marker='2')
ax.plot(x, x+17, color="blue", lw=1, ls='-', marker='3')
ax.plot(x, x+18, color="blue", lw=1, ls='-', marker='4')


# marker size and color
ax.plot(x, x+19, color="purple", lw=1, ls='-', marker='+', markersize=12, markeredgecolor='black')
ax.plot(x, x+20, color="purple", lw=1, ls='-', marker='o', markersize=7, markeredgecolor='black', markerfacecolor='yellow')
ax.plot(x, x+21, color="purple", lw=1, ls='-', marker=',', markersize=20, markeredgecolor='black')
ax.plot(x, x+22, color="purple", lw=1, ls='-', marker='.', markersize=15, markeredgecolor='black', markeredgewidth=2, markerfacecolor='yellow')
ax.plot(x, x+23, color="purple", lw=1, ls='-', marker='s', markersize=7, markeredgecolor='black', markeredgewidth=2, markerfacecolor='yellow')
ax.plot(x, x+24, color="purple", lw=1, ls='-', marker='1', markersize=12, markeredgecolor='black')
ax.plot(x, x+25, color="purple", lw=1, ls='-', marker='2', markersize=12, markeredgecolor='black')
ax.plot(x, x+26, color="purple", lw=1, ls='-', marker='3', markersize=12, markeredgecolor='black')
ax.plot(x, x+27, color="purple", lw=1, ls='-', marker='4', markersize=12, markeredgecolor='cyan')
```

### Control over axis appearance

In this section we will look at controlling axis sizing properties in a matplotlib figure.

## Plot range

We can configure the ranges of the axes using the `set_ylim` and `set_xlim` methods in the axis object, or `axis('tight')` for automatically getting "tightly fitted" axes ranges:

```python
fig,ax = plt.subplots(1,3,figsize=(12,4))

ax[0].plot(x, x**2, x, x**3)
ax[0].set_title('default axes')


# axes tight
ax[1].plot(x, x**2, x, x**3, color="purple", lw=2, ls='-.')
ax[1].axis('tight')
ax[1].set_title('axis tight')


# limit how much of axes is visibile
ax[2].plot(x, x**2, x, x**3)
ax[2].set_title('axes limit')
ax[2].set_xlim([0,1])
ax[2].set_ylim([0,2])


plt.tight_layout() # to stop overlapping
```
