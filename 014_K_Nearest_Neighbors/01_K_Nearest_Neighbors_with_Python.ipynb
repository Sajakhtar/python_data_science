# K Nearest Neighbors (KNN) Theory

Review Chapter 4 of Introduction to Statistical Learning (Gareth James et al.)

http://faculty.marshall.usc.edu/gareth-james/ISL/ISLR%20Seventh%20Printing.pdf


* K Nearest Neighbors (KNN) is a classification algorithm that operates on a very simple principle
* Classifying a data point based on labels (class) of nearest data points
* Training Algorithm
    1. Stora all the data
* Prediction Algorithm:
    1. Calculate the distance from x (new data point) to all points in your data
    2. Sort the points in your data by increasing the distance from x
    3. Predict the majority label (class) of the 'K' closest points
* choosing a K will affect what class a new point is assigned to
* Larger K smooths the boundry and creates more bias, with a cleaner cutoff at the cost of mislabeling some points
* Pros of KNN
    * very simple
    * Training is trivial
    * Works with any number of classes
    * Easy to add more data
    * Few Parameters
        * K = number of nearest neighbors
        * Distance Metric
* Cons
    * High Prediction Cost (worse for large data sets)
    * Not good with high dimensional (features) data
    * Categorical Features don't work well
* Common Data Science interview question
    * classify anonymous data using KNN



# K Nearest Neighbors with Python

You've been given a classified data set from a company! They've hidden the feature column names but have given you the data and the target classes. 

We'll try to use KNN to create a model that directly predicts a class for a new data point based off of the features.

Let's grab it and use it!

## Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Get the Data

Set index_col=0 to use the first column as the index.

```python
df = pd.read_csv('Classified Data', index_col=0)
```

```python
df.head()
```

```python
df.info()
# no categorical features, all continuous
```

```python
df.describe()
```

## Standardize the Variables

Because the KNN classifier predicts the class of a given test observation by identifying the observations that are nearest to it, the scale of the variables matters. Any variables that are on a large scale will have a much larger effect on the distance between the observations, and hence on the KNN classifier, than variables that are on a small scale.

```python
from sklearn.preprocessing import StandardScaler
```

```python
# create instance of standard scaler
scaler = StandardScaler()
```

```python
# fit the scaler to features
scaler.fit(df.drop('TARGET CLASS',axis=1))
```

```python
# transform method centers and scales features data using the fitted scaler object
scaled_features = scaler.transform(df.drop('TARGET CLASS',axis=1))
```

```python
# scaled_features is now an array or arays, scaled version of the features in original dataframe
scaled_features
```

```python
df.columns
```

```python
df.columns[:-1]
```

```python
# create a dataframe from scaled_features
df_feat = pd.DataFrame(scaled_features, columns=df.columns[:-1])
```

```python
# Now we have a standarized version of our data
df_feat.head()
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_feat # of X = scaled_features
```

```python
y = df['TARGET CLASS']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

## Using KNN

Remember that we are trying to come up with a model to predict whether someone will TARGET CLASS or not. We'll start with k=1, then use Elbow method to choose a K value.

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# create instance of KNN, specify K (number of neighbors)
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
# fit knn to training data
knn.fit(X_train, y_train)
```

## Predictions and Evaluations

Let's evaluate our KNN model!

```python
# Predictions
pred = knn.predict(X_test)
```

```python
pred[:10]
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
# Confusion Matrix
print(confusion_matrix(y_test, pred))
```

```python
# Classification Report
print(classification_report(y_test, pred))
# model is pretty good already with accuracy and other metrics in 90%+
# next we'll see if we can squeeze even more by choosing a better K value
```

## Choosing a K Value

Let's go ahead and use the elbow method to pick a good K Value using Elbow method:

```python
# iterate many models using diff K values, plot their error rates
# model with lowest error rate has best K value
error_rate = []

# will take some time
for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(X_test)
    error_i = np.mean(pred_i != y_test) # avg error rate
    error_rate.append(error_i) # append to list to plot later
```

```python
error_rate[:5]
```

```python
# plot the error rate
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate, '--', color='gray', marker='o', markerfacecolor='darkblue', markeredgecolor='darkred', markeredgewidth=2, markersize=7)
# linestyle='dashed'
plt.title('KNN: Error vs K')
plt.xlabel('K')
plt.ylabel('Err')
```

Here we can see that that after around K>23 the error rate just tends to hover around 0.06-0.05.
Let's retrain the model with that and check the classification report!
* K = 17 looks about right.
* K in 34+ looks dubious as it is bouncing around and our model was already good at K=1

```python
# K= 1, Accuracy 92%
# TN, FN
# FP, TP
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
pred = knn.predict(X_test)

print('WITH K=1')
print('\n')
print(confusion_matrix(y_test, pred))
print('\n')
print(classification_report(y_test, pred))
```

```python
# K= 17, Accuracy 95%, squeezed out a 6 more TP
# TN, FN
# FP, TP
knn = KNeighborsClassifier(n_neighbors=17)
knn.fit(X_train, y_train)
pred = knn.predict(X_test)

print('WITH K=17')
print('\n')
print(confusion_matrix(y_test, pred))
print('\n')
print(classification_report(y_test, pred))
```

We were able to squeeze some more performance out of our model by tuning to a better K value!
